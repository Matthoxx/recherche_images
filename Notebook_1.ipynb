```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from numpy import linalg as LA
from IPython.display import Image
SIZE=256

img = cv.imread('Capture.png', cv.IMREAD_COLOR)
Image(filename='Capture.png')
```

```python
green_histo = cv.calcHist( [ img ] , [ 1 ] , None, [256] , [0 , 256])
red_histo = cv.calcHist([ img ],[0], None, [256], [0,256])
blue_histo = cv.calcHist([ img ],[2], None, [256], [0,256])
fig, (ax1, ax2,ax3) = plt.subplots(1, 3)
fig.suptitle('RGB Hisotgrams of image')
ax1.plot(green_histo,color='g')
ax2.plot(red_histo, color='r')
ax3.plot(blue_histo, color='b')
fig.tight_layout()
```

```python
red_vector = red_histo.flatten()
green_vector = red_histo.flatten()
blue_vector = red_histo.flatten()
cv.normalize(red_vector, red_vector)
cv.normalize(green_vector, green_vector)
cv.normalize(blue_vector, blue_vector)
final_vector=np.concatenate((red_vector, green_vector, blue_vector)) 
plt.plot(final_vector)
plt.show()
```
